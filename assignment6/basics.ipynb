```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print(3.14 * self.r * self.r)

    def perimeter(self):
        print(2 * 3.14 * self.r)


c = Circle(5)
c.area()
c.perimeter()
```

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        return a / b


c = Calculator()

print("Addition:", c.add(10, 5))
print("Subtraction:", c.subtract(10, 5))
print("Multiplication:", c.multiply(10, 5))
print("Division:", c.divide(10, 5))
```
