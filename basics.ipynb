```python
numbers = [1, 2, 2, 3, 3, 4, 4, 4, 4]
frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

print(frequency)
```

```python
sentence = input("Enter a sentence: ")
words = sentence.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
students = {
    "Ali": 85,
    "Sara": 92,
    "Ahmed": 78,
    "Zara": 95
}

highest_student = ""
highest_score = 0

for name, score in students.items():
    if score > highest_score:
        highest_score = score
        highest_student = name

print("Highest scoring student:", highest_student, "with", highest_score, "marks")
```
