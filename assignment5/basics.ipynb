```python
numbers = [5, 12, 7, 20, 3, 8, 15, 10]

print("All numbers in the list:")
for i in numbers:
    print(i, end=" ")
print()

total = 0
for i in numbers:
    total += i
print("Sum of all numbers:", total)

print("Even numbers in the list:")
for i in numbers:
    if i % 2 == 0:
        print(i, end=" ")
print()
```

```python
student = {
    "name": "Ali",
    "age": 20,
    "course": "Python",
    "marks": [60, 45, 70]
}

print("Student Details:")
for key, value in student.items():
    print(f"{key}: {value}")


sum_marks = 0
for i in student["marks"]:
    sum_marks += i
average = sum_marks / len(student["marks"])
print("Average Marks:", average)


if average >= 50:
    print("Result: Passed")
else:
    print("Result: Failed")
```

```python
employees = [
    {"name": "Ahmed", "department": "IT", "salary": 50000},
    {"name": "Sara", "department": "HR", "salary": 60000},
    {"name": "Bilal", "department": "Finance", "salary": 55000}
]


print("Employee Details:")
for i in employees:
    print(i)


highest_salary = employees[0]
for i in employees:
    if i["salary"] > highest_salary["salary"]:
        highest_salary = i
print("Highest Salary:", highest_salary)


total_salary = 0
for i in employees:
    total_salary += i["salary"]
print("Total Salary Expense:", total_salary)
```

```python
user_numbers = []
print("Enter numbers (-1 to stop):")
while True:
    num = int(input())
    if num == -1:
        break
    user_numbers.append(num)

print("Numbers entered:", user_numbers)
if user_numbers:  
    print("Largest number:", max(user_numbers))
    print("Smallest number:", min(user_numbers))
```

```python
sentence = input("Enter a sentence: ")
words = sentence.split() 
word_count = {}


for word in words:
    word = word.lower()  
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1


print("Word Frequency:")
for word, count in word_count.items():
    print(f"{word}: {count}")
```
